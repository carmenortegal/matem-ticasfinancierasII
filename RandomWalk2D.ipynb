## Random Walk en 2D

```python
#En primer lugar, se instalan las librerías que nos permitirán realizar los diferentes cálculos necesarios
import random  #generación de números aleatorios
import math  #cálculo de diferentes operaciones matemáticas 
import matplotlib.pyplot as plt  #generación de gráficos
```

```python
random.seed() 
n=1000 # número de pasos  
r=1 # longitud del paso  
x=y=0 # punto de partida (x,y)=(0,0)  
listax=[x]  #Se genera una lista que recogerá todos los valores de x 
listay=[y]  #Se genera una lista que recogerá todos los valores de y
for i in range(n):  
  angulo=random.random()*math.pi*2 #Cálculo del ángulo del siguiente movimiento en radianes, mediante la generación de números aleatorios
  x+=r*math.cos(angulo)  
  y+=r*math.sin(angulo)  
  listax.append(x)  #se concatena cada valor de x a la lista existente
  listay.append(y)  #se concatena cada valor de y a la lista existente
plt.plot(listax,listay)  
plt.show() 

#Adicionalmente, se podrían mostrar los valores que toman x e y 
print(f"Valores de X: {listax}")
print(f"Valores de Y: {listay}")
```
